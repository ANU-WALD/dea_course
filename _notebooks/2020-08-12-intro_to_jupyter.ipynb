#### Welcome to the Jupyter notebook environment. This is an interactive way to run Python code in your browser. The gray cell below contains Python code that you can run placing the cursor on the cell and clicking on the triangle icon on the top menu.

#### Tip: _Use the shortcut `Shift-Enter` for running the selected cell, it's much more convenient than clicking the icon each time._

````python
print("Welcome to the DEA Sandbox!")
````

#### You'll see that, once the code has executed, any output generated by your code will show up just below the cell

````python
print("3,2,1")

# Use hash for commenting your code

print("this is fun!")
````

#### The code that you execute using the Sandbox does not run on your local laptop. It runs in a remote server and you just receive the output of the code.

#### Tip: _The last line of code in a cell is printed by default, so there is no need to type `print()`_

````python
# Sending a complicated operation to the server
result = 999 + 1

# This operation goes to the server and we just display its output in our screens once is completed
result
````

#### Sometimes the code that you run takes a while to compute because it requires to perform complex computations. You'll notice the number that appears next to the cell. Before the cell is run you'll see the symbol `[ ]`, while the cell is being run it shows as `[*]` and once it has completed running you'll see a number representing the number of cells being run, for example `[4]`.

````python
# This is how a library is imported in Python but don't worry about this for the moment:
import time

# Very complicated and long operation computed on the server
time.sleep(5)

print("Done!")
````
